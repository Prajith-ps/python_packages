```python
#importing dataset
import pandas as pd
df=pd.read_csv('cardata.csv')
df.head()
```

```python
df.describe
```

```python
# scaling dataset
from sklearn.preprocessing import MinMaxScaler
numeric_cols=['Year','Selling_Price','Present_Price','Kms_Driven','Owner']
scaler=MinMaxScaler()
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])
df[numeric_cols].head()
```

```python
from sklearn.model_selection import train_test_split
X=df[['Year','Kms_Driven']]
y=df[['Selling_Price']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
print('x_train \n',X_train.head(),
      '\n x_test \n',X_test.head(),
      '\n y_train \n',y_train.head(),
      '\n y_test \n',y_test.head())
```

```python
#making prediction
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
new_data=[[0.33423,0.093498]]
pred_value=model.predict(new_data)
pred_value
```

```python
#importing dataset
import pandas as pd
df=pd.read_csv('salary_dataset.csv')
df.head()
```

```python
df.describe
```

```python
# scaling dataset
from sklearn.preprocessing import MinMaxScaler
numeric_cols=['YearsExperience','Salary',]
scaler=MinMaxScaler()
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])
df[numeric_cols].head()
```

```python
from sklearn.model_selection import train_test_split
X=df[['YearsExperience']]
y=df[['Salary']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
print('x_train \n',X_train.head(),
      '\n x_test \n',X_test.head(),
      '\n y_train \n',y_train.head(),
      '\n y_test \n',y_test.head())
```

```python
#making prediction
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
new_data=[[0.33423]]
pred_value=model.predict(new_data)
pred_value
```
