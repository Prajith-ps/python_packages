```python
import pandas as pd
data1={'Name':['prajith','rahul','arjun','unni'],
       'age':[20,21,22,23],
       'address':['kochi','kottayam','kannur','kozhi'],
       'qualification':['it','bsc','b.com','b.tech']}
data2={'Name':['adhi','kiran','anandhu','fazeem'],
       'age':[20,30,33,24],
       'address':['kochi','chennai','goa','kottayam'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1,index=[0,1,2,3,])
df1=pd.DataFrame(data2,index=[1,2,3,4])
print(df,'\n\n',df1)
```

```python
#concatination
rslt=pd.concat([df,df1])
rslt
```

```python
#setting logic on axis
rslt1=pd.concat([df,df1],axis=1,join='inner')
rslt1
```

```python
#outer join
rslt1=pd.concat([df,df1],axis=1,sort=False)
rslt1
```

```python
#append
rslt=df.append(df1)
rslt
```

```python
#merging

import pandas as pd
data1={'id':['d0','d1','d2','d3'],
       'Name':['prajith','rahul','arjun','unni'],
       'age':[20,21,22,23]}
data2={'id':['d1','d2','d3','d4'],
       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[1,2,3,4])
print(df,'\n\n',df1)
```

```python
#merging
rslt=pd.merge(df,df1,on="id")
rslt
```

```python
#joinig with multiple keys
import pandas as pd
data1={'id':['d0','d1','d2','d3'],
       'id1':['d0','d1','d1','d0'],
       'Name':['prajith','rahul','arjun','unni'],
       'age':[20,21,22,23]}
data2={'id':['d1','d2','d3','d4'],
       'id1':['d0','d1','d0','d1'],
       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,'\n\n',df1)
```

```python
rslt=pd.merge(df,df1,on=['id','id1'])
rslt
```

```python
#how=left
rslt=pd.merge(df,df1,how='left',on=['id','id1'])
rslt
```

```python


#how=right
rslt=pd.merge(df,df1,how='right',on=['id','id1'])
rslt

```

```python
#how=outer
rslt=pd.merge(df,df1,how='outer',on=['id','id1'])
rslt
```

```python
#how=inner
rslt=pd.merge(df,df1,how='inner',on=['id','id1'])
rslt
```
