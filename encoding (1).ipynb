```python
import pandas as pd
#one-hot encoding
df=pd.DataFrame({'colors':['red','green','blue','orange','yellow']})
print('Dataframe before encoding')
print(df['colors'])
one_hot=pd.get_dummies(df['colors'])
# df1=pd.concat([df,one_hot],axis=1)
# df1=df1.drop('colors',axis=1)
print('Dataframe after encoding')
one_hot
```

```python
#lable encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print('Befor encoding')
print(df['colors'])
df['color_label']=le.fit_transform(df['colors'])
print('After encoding')
df
```

```python
#label encoding
df=pd.read_csv('cardata.csv')
ctgry_cols=['Fuel_Type','Seller_Type','Transmission']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for cols in ctgry_cols:
    df[cols]=le.fit_transform(df[cols])
print(df.head())
```

```python
#minmaxscalar

from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
numeric_cols=['Year','Selling_Price','Present_Price','Kms_Driven','Owner']
df[numeric_cols]=scalar.fit_transform(df[numeric_cols])
df.head()
```

```python
scalar=MinMaxScaler()
numeric_cols=['Year','Selling_Price','Present_Price','Kms_Driven','Owner']
scaled_data=scalar.fit_transform(df[numeric_cols])
scaled_df=pd.DataFrame(scaled_data,columns=numeric_cols)
df.head()
```

```python
# normalization
from sklearn.preprocessing import Normalizer
scaler=Normalizer()
numeric_cols=['Year','Selling_Price','Present_Price','Kms_Driven','Owner']
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])
df.head()
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numeric_cols=['Year','Selling_Price','Present_Price','Kms_Driven','Owner']
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])
df.head()
```
